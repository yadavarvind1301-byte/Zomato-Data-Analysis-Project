```python
import pandas as pd
import seaborn as sns

dataframe = pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

```python
#convert the data type of column -rate
```

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
dataframe['rate'] = dataframe['rate'].astype(str).str.split('/').str[0]
dataframe['rate'] = pd.to_numeric(dataframe['rate'], errors='coerce')
```

```python
dataframe.info()
```

```python
#Type of resturant
```

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("type of resturent")
```

Conclusion- majorityof the resturant falls in dinning category

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result, c="green",marker="o")
plt.xlabel("Type of restaurent", c="red", size=20)
plt.ylabel("votes",c="red",size=20)
```

conclusion-- dinning resturants has recieved maximum votes

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins = 2)
plt.title("ratings distribustion")
plt.show()
```

conclusion-- The majority resturants received ratings from 3.5 to 4 

```python
dataframe.head()
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

```python
dataframe.head()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)
```

Conculision-- offline order recived lower rating in comparison to online order

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(
    index='listed_in(type)',
    columns='online_order',
    aggfunc='size',
    fill_value=0
)

sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed Type")
plt.show()
```

CONCULISION: Dining restaurants primarily accept offline orders, whereas cafes primarily receive online oreder. this suggestts that clients prefer to place oredr in person at restaurants,but prefer online ordering at cafes.
